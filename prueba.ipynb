```python
import pandas as pd

file_path = r'C:\Users\nekan\Desktop\HUS\2025-HOSPITAL\prueba5enf.xlsx'
df = pd.read_excel(file_path)
df.head()
```

```python
df = df.drop(index=0)  
df.reset_index(drop=True, inplace=True) 
df.head()
```

```python
first_column_values = df.iloc[:, 0].values
print(first_column_values)
```

```python
df.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
df.head()
```

```python
new_column_names = {df.columns[i]: i-1 for i in range(1, len(df.columns))}
df.rename(columns=new_column_names, inplace=True)
df.head()
```

```python
id_categories = {}

for column in df.columns[1:]:  # Skip the 'ID' column
    id_categories[column] = {'M': [], 'T': [], 'N': []}
    for idx, value in enumerate(df[column]):
        if value in ['M', 'T', 'N']:
            id_categories[column][value].append(df['ID'][idx])

print(id_categories)
```
